```python
import pandas as pd
import numpy as np



# 2. define alpha (w1) search space
w1_candidates = np.linspace(0, 1, 11) # 0.0, 0.1, ..., 1.0

print(f"{'w1 (Weight)':<15} | {'r with Consistency':<20} | {'r with Anomaly Rate':<20}")
print("-" * 60)

for w1 in w1_candidates:
    # 3. Formula：Trust = w1 * (1 - AR) + (1 - w1) * CS
    df['temp_trust'] = w1 * (1 - df['anomaly_rate']) + (1 - w1) * df['consistency']
    
    # 4. calculate correlation r
    # observe Trust Score with other two factors correlation 
    r_c = df['temp_trust'].corr(df['consistency'])
    r_a = df['temp_trust'].corr(df['anomaly_rate'])
    
    print(f"{w1:<15.1f} | {r_c:<20.4f} | {r_a:<20.4f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. sucess case study LSTM machine-1-1
scores = np.load("artifacts/scores/machine-1-1/LSTM_scores.npy")
total_len = len(scores)

# 2. explain SMD interpretation label
def get_labels_from_interpretation(file_path, length):
    label_arr = np.zeros(length, dtype=int)
    with open(file_path, 'r') as f:
        for line in f:
            line = line.strip()
            if not line:
                continue
            interval = line.split(':')[0]
            start, end = map(int, interval.split('-'))
            label_arr[start:end+1] = 1
    return label_arr

label_path = "ServerMachineDataset/interpretation_label/machine-1-1.txt"
labels = get_labels_from_interpretation(label_path, total_len)

# 3. offline PA logic
threshold = 0.048591
y_pred_raw = (scores > threshold).astype(int)
y_pred_pa = y_pred_raw.copy()

# find anomaly segments
anomaly_segments = np.where(np.diff(np.pad(labels, (1, 1), mode='constant')) != 0)[0].reshape(-1, 2)

# PA complete
for start, end in anomaly_segments:
    # end is the right boundary of the open interval
    if np.any(y_pred_raw[start:end]):
        y_pred_pa[start:end] = 1

# 4.  exact only key area
# take all anomaly range，add vor/afterwards buffer
buffer = 1500

if len(anomaly_segments) > 0:
    focus_start = max(0, anomaly_segments[0, 0] - buffer)
    focus_end = min(total_len, anomaly_segments[-1, 1] + buffer)
else:
    focus_start = 0
    focus_end = total_len

scores_plot = scores[focus_start:focus_end]
labels_plot = labels[focus_start:focus_end]
y_pred_pa_plot = y_pred_pa[focus_start:focus_end]
x_plot = np.arange(focus_start, focus_end)


score_cap = np.percentile(scores_plot, 99.5)
scores_vis = np.clip(scores_plot, None, score_cap)

# 5. plot
fig, (ax1, ax2) = plt.subplots(
    2, 1, figsize=(12, 6), sharex=True,
    gridspec_kw={'height_ratios': [2.5, 1]}
)

# upper panel：Score + threshold + GT
ax1.plot(x_plot, scores_vis, color='royalblue', label='LSTM Score', alpha=0.85, linewidth=1.2)
ax1.axhline(y=threshold, color='red', linestyle='--', linewidth=1.2, label=f'Threshold ({threshold:.6f})')

# fill_between mark anomaly area
ax1.fill_between(
    x_plot, 0, score_cap,
    where=(labels_plot == 1),
    color='green', alpha=0.12, label='Ground Truth (GT)'
)

ax1.set_title('Offline Evaluation (with PA): Machine-1-1 LSTM', fontsize=13)
ax1.set_ylabel('Anomaly Score')
ax1.set_ylim(0, score_cap * 1.05)
ax1.legend(loc='upper right')

# lower panel：GT vs PA-adjusted prediction
ax2.fill_between(x_plot, 0, labels_plot, color='green', alpha=0.20, label='Ground Truth')
ax2.step(x_plot, y_pred_pa_plot, where='post', color='darkorange',
         linewidth=1.4, label='Final Prediction (PA Adjusted)')

ax2.set_yticks([0, 1])
ax2.set_yticklabels(['Normal', 'Anomaly'])
ax2.set_xlabel('Time Steps')
ax2.set_ylabel('State')
ax2.legend(loc='upper right')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# =========================
# 1. Load data
# =========================
scores = np.load("artifacts/scores/machine-1-2/LSTM_scores.npy")
total_len = len(scores)

def get_labels_from_interpretation(file_path, length):
    label_arr = np.zeros(length, dtype=int)
    with open(file_path, "r") as f:
        for line in f:
            line = line.strip()
            if not line:
                continue
            interval = line.split(":")[0]
            start, end = map(int, interval.split("-"))
            label_arr[start:end + 1] = 1
    return label_arr

label_path = "ServerMachineDataset/interpretation_label/machine-1-2.txt"
labels = get_labels_from_interpretation(label_path, total_len)

threshold = 0.00035254325019195816

# =========================
# 2. Raw prediction + Point Adjustment
# =========================
y_pred_raw = (scores > threshold).astype(int)
y_pred_pa = y_pred_raw.copy()

anomaly_segments = np.where(
    np.diff(np.pad(labels, (1, 1), mode="constant")) != 0
)[0].reshape(-1, 2)

for start, end in anomaly_segments:
    if np.any(y_pred_raw[start:end]):
        y_pred_pa[start:end] = 1

# =========================
# 3. Print anomaly segments
# =========================
print("Anomaly segments:")
for i, (s, e) in enumerate(anomaly_segments):
    print(f"{i}: {s} - {e}")

# =========================
# 4. Unified local zoom plot
# =========================
def plot_local_zoom(segment_idx, pad=250, title_prefix="Machine-1-2 LSTM", save_path=None):
    start, end = anomaly_segments[segment_idx]
    left = max(0, start - pad)
    right = min(total_len, end + pad)

    x_plot = np.arange(left, right)
    scores_plot = scores[left:right]
    labels_plot = labels[left:right]
    pred_plot = y_pred_pa[left:right]

    ymax = max(scores_plot.max(), threshold * 1.2)

    fig, (ax1, ax2) = plt.subplots(
        2, 1,
        figsize=(10, 4.8),
        sharex=True,
        gridspec_kw={"height_ratios": [2.3, 1]}
    )

    # -------------------------
    # Upper panel: score + threshold + GT
    # -------------------------
    ax1.plot(
        x_plot,
        scores_plot,
        color="royalblue",
        linewidth=1.1,
        alpha=0.95,
        label="Anomaly Score"
    )

    ax1.axhline(
        y=threshold,
        color="red",
        linestyle="--",
        linewidth=1.1,
        label=f"Threshold ({threshold:.6f})"
    )

    ax1.fill_between(
        x_plot,
        0,
        ymax * 1.02,
        where=(labels_plot == 1),
        color="green",
        alpha=0.12,
        label="Ground Truth (GT)"
    )

    ax1.set_ylabel("Anomaly Score")
    ax1.set_ylim(0, ymax * 1.05)
    ax1.set_title(f"Zoom in anomaly segment {segment_idx + 1}: {title_prefix}", fontsize=12)
    ax1.legend(loc="upper right")

    # -------------------------
    # Lower panel: GT vs Prediction (same layer)
    # -------------------------
    ax2.fill_between(
        x_plot, 0, labels_plot,
        color="green",
        alpha=0.20,
        label="Ground Truth"
    )

    ax2.fill_between(
        x_plot, 0, pred_plot,
        color="darkorange",
        alpha=0.25,
        label="Final Prediction (PA Adjusted)"
    )

    ax2.set_ylim(-0.05, 1.05)
    ax2.set_yticks([0, 1])
    ax2.set_yticklabels(["Normal", "Anomaly"])
    ax2.set_xlabel("Time Steps")
    ax2.set_ylabel("State")
    ax2.legend(loc="upper right")

    plt.tight_layout()

    if save_path:
        plt.savefig(save_path, dpi=300, bbox_inches="tight")

    plt.show()

# =========================
# 5. Example usage
# =========================
plot_local_zoom(segment_idx=8, pad=250)
plot_local_zoom(segment_idx=4, pad=250)
```
